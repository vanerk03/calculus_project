```python
from generate import generate
import numpy as np
from model import LinearRegressor

system = np.array([[1, 2], [-3, -5]])
X, y = generate(system, dataset_size=500, scope=12)
```

\begin{equation*}
 \begin{cases}
   x'(t) = 1x + 2y\\
   y'(t) = -3x - 5y
 \end{cases}
\end{equation*}

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def f(s, t):
    x = s[0]
    y = s[1]
    dydt = 1 * x + 2 * y
    dxdt = -3 * x - 5 * y
    return [dydt, dxdt]

t = np.linspace(0,20)
s0=[20,5]

s = odeint(f, s0, t)

plt.plot(t,s[:,0],'r--', linewidth=2.0)
plt.plot(t,s[:,1],'b-', linewidth=2.0)
plt.xlabel("t")
plt.ylabel("S[N,C]")
plt.legend(["N","C"])
plt.show()
```
